2) AIM: Write a program that prompts a user to enter element of a list and add the elements to a
list. Write a function maxim(list) and minum(list) to find out the maximum and minimum values
from list.


```python
list=[]
n=int(input())
for i in range(n):
  m=int(input())
  list.append(m)
print(list)
def maximum(list):
  a=max(list)
  return a
def minimum(list):
  b=min(list)
  return b
print(maximum(list))
print(minimum(list))
```

3) AIM: Write a function reverse(list) to reverse the elements of the list

```python
def reverse(b):
  b.reverse()
  return b
b=[1,7,3]
print(reverse(b))
```

4) AIM: Write a program to return prime numbers from a list.


```python
l=[int(i) for i in input().split()]
c=0
for i in l:
  for j in range(2,i):
    if(i%j==0):
      c=c+1
  if(c==0):
      print(i)
```
